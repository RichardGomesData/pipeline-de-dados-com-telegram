## 1\. Ingesto

A etapa de **ingesto**  responsvel, como seu o prprio nome diz, pela ingesto dos dados transacionais em ambientes analticos. De maneira geral, o dado ingerido  persistido no formato mais prximo do original, ou seja, nenhuma transformao  realizada em seu contedo ou estrutura (*schema*). Como exemplo, dados de uma API *web* que segue o formato REST (*representational state transfer*) so entregues, logo, persistidos, no formato JSON.

> Persistir os dados em seu formato original trs muitas vantagens, como a possibilidade de reprocessamento.

Pode ser conduzida de duas formas:

 - **Batch**: blocos de dados so ingeridos em uma frequncia bem definida, geralmente na escala de horas ou dias;
 - **Streaming**: dados so ingeridos conforme so produzidos e disponibilizados.

No projeto, as mensagens capturadas pelo *bot* podem ser ingeridas atravs da API *web* de *bots* do **Telegram**, portanto so fornecidos no formato JSON. Como o **Telegram** retem mensagens por apenas 24h em seus servidores, a ingesto via **streaming**  a mais indicada. Para que seja possvel esse tipo de **ingesto** seja possvel, vamos utilizar um *webhook* (gancho *web*), ou seja, vamos redirecionar as mensagens automaticamente para outra API *web*.

Sendo assim, precisamos de um servio da AWS que fornea um API *web* para receber os dados redirecionados, o `AWS API Gateway` (documentao neste [link](https://docs.aws.amazon.com/pt_br/apigateway/latest/developerguide/welcome.html)). Dentre suas diversas funcionalidades, o `AWS API Gateway` permite o redirecionamento do dado recebido para outros servios da AWS. Logo, vamos conecta-lo ao `AWS Lambda`, que pode sua vez, ir armazenar o dado em seu formato original (JSON) em um *bucket* do `AWS S3`.

> Sistemas que reagem a eventos so conhecidos como *event-driven*.

Portanto, precisamos:

 - Criar um *bucket* no `AWS S3`;
 - Criar uma funo no `AWS Lambda`;
 - Criar uma API *web* no `AWS API Gateway`;
 - Configurar o *webhook* da API de *bots* do **Telegram**.

### **1.1. AWS S3**

Na etapa de **ingesto**, o `AWS S3` tem a funo de passivamente armazenar as mensagens captadas pelo *bot* do **Telegram** no seu formato original: JSON. Para tanto, basta a criao de um *bucket*. Como padro, vamos adicionar o sufixo `-raw` ao seu nome (vamos seguir esse padro para todos os servios desta camada).

> **Nota**: um `data lake`  o nome dado a um repositrio de um grande volume dados.  organizado em zonas que armazenam replicadas dos dados em diferentes nveis de processamento. A nomenclatura das zonas varia, contudo, as mais comuns so: *raw* e *enriched* ou *bronze*, *silver* e *gold*.

### **1.2. AWS Lambda**

Na etapa de **ingesto**, o `AWS Lambda` tem a funo de ativamente persistir as mensagens captadas pelo *bot* do **Telegram** em um *bucket* do `AWS S3`. Para tanto vamos criar uma funo que opera da seguinte forma:

 - Recebe a mensagem no parmetro `event`;
 - Verifica se a mensagem tem origem no grupo do **Telegram** correto;
 - Persiste a mensagem no formato JSON no *bucket* do `AWS S3`;
 - Retorna uma mensagem de sucesso (cdigo de retorno HTTP igual a 200) a API de *bots* do **Telegram**.

> **Nota**: No **Telegram**, restringimos a opo de adicionar o *bot* a grupos, contudo, ainda  possvel iniciar uma conversa em um *chat* privado.

O cdigo da funo:

````
import os
import json
import logging
from datetime import datetime, timezone

import boto3


def lambda_handler(event: dict, context: dict) -> dict:

  '''
  Recebe uma mensagens do Telegram via AWS API Gateway, verifica no
  seu contedo se foi produzida em um determinado grupo e a escreve,
  em seu formato original JSON, em um bucket do AWS S3.
  '''

  # vars de ambiente

  BUCKET = os.environ['AWS_S3_BUCKET']
  TELEGRAM_CHAT_ID = int(os.environ['TELEGRAM_CHAT_ID'])

  # vars lgicas

  tzinfo = timezone(offset=timedelta(hours=-3))
  date = datetime.now(tzinfo).strftime('%Y-%m-%d')
  timestamp = datetime.now(tzinfo).strftime('%Y%m%d%H%M%S%f')

  filename = f'{timestamp}.json'

  # cdigo principal

  client = boto3.client('s3')

  try:

    message = json.loads(event["body"])
    chat_id = message["message"]["chat"]["id"]

    if chat_id == TELEGRAM_CHAT_ID:

      with open(f"/tmp/{filename}", mode='w', encoding='utf8') as fp:
        json.dump(message, fp)

      client.upload_file(f'/tmp/{filename}', BUCKET, f'telegram/context_date={date}/{filename}')

  except Exception as exc:
      logging.error(msg=exc)
      return dict(statusCode="500")

  else:
      return dict(statusCode="200")
````

Note que o cdigo exige a configurao de duas variveis de ambiente: `AWS_S3_BUCKET` com o nome do *bucket* do `AWS S3` e `TELEGRAM_CHAT_ID` com o id do *chat* do grupo do **Telegram**. Para adicionar variveis de ambiente em uma funo do `AWS Lambda`, basta acessar configuraes -> variveis de ambiente no console da funo.

## 2\. ETL

A etapa de **extrao, transformao e carregamento** (do ingls *extraction, transformation and load* ou **ETL**)  uma etapa abrangente responsvel pela manipulao dos dados ingeridos de sistemas transacionais, ou seja, j persistidos em camadas cruas ou *raw* de sistemas analticos. Os processos conduzidos nesta etapa variam bastante de acordo com a rea da empresa, do volume/variedade/velocidade do dado consumido, etc. Contudo, em geral, o dado cru ingerido passa por um processo recorrente de *data wrangling* onde o dado  limpo, deduplicado, etc. e persistido com tcnicas de particionamento, orientao a coluna e compresso. Por fim, o dado processado est pronto para ser analisado por profissionais de dados.

No projeto, as mensagens de um nico dia, persistidas na camada cru, sero compactas em um nico arquivo, orientado a coluna e comprimido, que ser persistido em uma camada enriquecida. Alm disso, durante este processo, o dado tambm passar por etapas de *data wrangling*.

Para isso, vamos utilizar uma funo do `AWS Lambda` como motor de processamento e um *bucket* do `AWS S3` como camada enriquecida para a persistncia do dado processado. Para garantir a recorrncia, vamos configurar uma regra do `AWS Event Bridge` como gatilho diro da funo.

### **2.1. AWS S3**

Na etapa de **ETL**, o `AWS S3` tem a funo de passivamente armazenar as mensagens processadas de um dia em um nico arquivo no formato Parquet. Para tanto, basta a criao de um *bucket*. Como padro, vamos adicionar o sufixo `-enriched` ao seu nome (vamos seguir esse padro para todos os servios desta camada).

> **Nota**: um `data lake`  o nome dado a um repositrio de um grande volume dados.  organizado em zonas que armazenam replicadas dos dados em diferentes nveis de processamento. A nomenclatura das zonas varia, contudo, as mais comuns so: *raw* e *enriched* ou *bronze*, *silver* e *gold*.

### **2.2. AWS Lambda**

Na etapa de **ETL**, o `AWS Lambda` tem a funo de ativamente processar as mensagens captadas pelo *bot* do **Telegram**, persistidas na camada cru no *bucket* do `AWS S3`, e persisti-las na camada enriquecida, tambm em um *bucket* do `AWS S3`. Logo, vamos criar uma funo que opera da seguinte forma:

 - Lista todos os arquivos JSON de uma nica participao da camada crua de um *bucket* do `AWS S3`;
 - Para cada arquivo listado:
  - Faz o *download* do arquivo e carrega o contedo da mensagem;
  - Executa uma funo de *data wrangling*;
  - Cria uma tabela do PyArrow e a contatena com as demais.
 - Persiste a tabela no formato Parquet na camada enriquecida em um *bucket* do `AWS S3`.

> **Nota**: O fato de utilizarmos duas camadas de armazenamento e processamento, permite que possamos reprocessar os dados crus de diversas maneiras, quantas vezes forem preciso.

> **Nota**: Atente-se ao fato de que a funo processa as mensagens do dia anterior (D-1).

O cdigo da funo:

````
import os
import json
import logging
from datetime import datetime, timedelta, timezone

import boto3
import pyarrow as pa
import pyarrow.parquet as pq


def lambda_handler(event: dict, context: dict) -> bool:

  '''
  Diariamente  executado para compactar as diversas mensagensm, no formato
  JSON, do dia anterior, armazenadas no bucket de dados cru, em um nico
  arquivo no formato PARQUET, armazenando-o no bucket de dados enriquecidos
  '''

  # vars de ambiente

  RAW_BUCKET = os.environ['AWS_S3_BUCKET']
  ENRICHED_BUCKET = os.environ['AWS_S3_ENRICHED']

  # vars lgicas

  tzinfo = timezone(offset=timedelta(hours=-3))
  date = (datetime.now(tzinfo) - timedelta(days=1)).strftime('%Y-%m-%d')
  timestamp = datetime.now(tzinfo).strftime('%Y%m%d%H%M%S%f')

  # cdigo principal

  table = None
  client = boto3.client('s3')

  try:

      response = client.list_objects_v2(Bucket=RAW_BUCKET, Prefix=f'telegram/context_date={date}')

      for content in response['Contents']:

        key = content['Key']
        client.download_file(RAW_BUCKET, key, f"/tmp/{key.split('/')[-1]}")

        with open(f"/tmp/{key.split('/')[-1]}", mode='r', encoding='utf8') as fp:

          data = json.load(fp)
          data = data["message"]

        parsed_data = parse_data(data=data)
        iter_table = pa.Table.from_pydict(mapping=parsed_data)

        if table:

          table = pa.concat_tables([table, iter_table])

        else:

          table = iter_table
          iter_table = None

      pq.write_table(table=table, where=f'/tmp/{timestamp}.parquet')
      client.upload_file(f"/tmp/{timestamp}.parquet", ENRICHED_BUCKET, f"telegram/context_date={date}/{timestamp}.parquet")

      return True

  except Exception as exc:
      logging.error(msg=exc)
      return False
````

O cdigo da funo de *data wrangling*:

````
def parse_data(data: dict) -> dict:

  date = datetime.now().strftime('%Y-%m-%d')
  timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

  parsed_data = dict()

  for key, value in data.items():

      if key == 'from':
          for k, v in data[key].items():
              if k in ['id', 'is_bot', 'first_name']:
                parsed_data[f"{key if key == 'chat' else 'user'}_{k}"] = [v]

      elif key == 'chat':
          for k, v in data[key].items():
              if k in ['id', 'type']:
                parsed_data[f"{key if key == 'chat' else 'user'}_{k}"] = [v]

      elif key in ['message_id', 'date', 'text']:
          parsed_data[key] = [value]

  if not 'text' in parsed_data.keys():
    parsed_data['text'] = [None]

  return parsed_data
````

Para que a funo funcione corretamente, algumas configuraes precisam ser realizadas.

 - **Variveis de ambiente**

Note que o cdigo exige a configurao de duas variveis de ambiente: `AWS_S3_BUCKET` e `AWS_S3_ENRICHED` com os nomes dos *bucket* do `AWS S3` da camada cru e enriquecida, respectivamente. Para adicionar variveis de ambiente em uma funo do `AWS Lambda`, basta acessar configuraes -> variveis de ambiente no console da funo.

 - **Permisso**

Precisamos adicionar a permisso de escrita nos *buckets* do `AWS S3` para a funo do `AWS Lambda` no `AWS IAM`.

 - **Recursos**

O *timeout* padro de funces do `AWS Lambda`  de 3 segundos. Para a funo, vamos aumentar o tempo para 5 minutos, principalmente para lidar com o IO (*input/output*) de arquivos do `AWS S3`.

 - **Camadas**

Por fim, note que o cdigo da funo utiliza o pacote Python PyArrow. Contudo, o ambiente padro do `AWS Lambda` possui poucos pacotes externos instalado, como o pacote Python boto3, logo o PyArrow no ser encontrado e a execuo da funo falhar. Existem algumas formas de adicionar pacotes externos no ambiente de execuo do AWS Lambda, um deles  a criao de camadas ou *layers*, onde podemos fazer o *upload* dos pacotes Python direto na plataforma ou atravs de um *bucket* do `AWS S3`. Vamos ento seguir com a ltima opo, onde teremos que:

 - Criar um *bucket* no `AWS S3`;
 - Fazer o *upload* do cdigo do pacote Python do PyArrow (*download* neste [link](https://github.com/awslabs/aws-data-wrangler/releases));
 - Criar *layer* e conectar na funo.

### **2.3. AWS Event Bridge**

Na etapa de **ETL**, o `AWS Event Bridge` tem a funo de ativar diariamente a funo de **ETL** do `AWS Lambda`, funcionando assim como um *scheduler*.

## 3\. Apresentao

A etapa de **apresentao**  reponsvel por entregar o dado para os usurios (analistas, cientistas, etc.) e sistemas (dashboards, motores de consultas, etc.), idealmente atravs de uma interface de fcil uso, como o SQL, logo, essa  a nica etapa que a maioria dos usurios ter acesso. Alm disso,  importante que as ferramentas da etapa entregem dados armazenados em camadas refinadas, pois assim as consultas so mais baratas e o dados mais consistentes.

### **3.1. AWS Athena**

Na etapa de **apresentao**, o `AWS Athena` tem funo de entregar o dados atravs de uma interface SQL para os usurios do sistema analtico. Para criar a interface, basta criar uma tabela externa sobre o dado armazenado na camada mais refinada da arquitetura, a camada enriquecida.

```sql
CREATE EXTERNAL TABLE `telegram`(
  `message_id` bigint,
  `user_id` bigint,
  `user_is_bot` boolean,
  `user_first_name` string,
  `chat_id` bigint,
  `chat_type` string,
  `text` string,
  `date` bigint)
PARTITIONED BY (
  `context_date` date)
ROW FORMAT SERDE
  'org.apache.hadoop.hive.ql.io.parquet.serde.ParquetHiveSerDe'
STORED AS INPUTFORMAT
  'org.apache.hadoop.hive.ql.io.parquet.MapredParquetInputFormat'
OUTPUTFORMAT
  'org.apache.hadoop.hive.ql.io.parquet.MapredParquetOutputFormat'
LOCATION
  's3://<bucket-enriquecido>/'
```

parsed_date_cte AS (
    SELECT
        *,
        CAST(date_format(from_unixtime("date"),'%Y-%m-%d %H:%i:%s') AS timestamp) AS parsed_date

    parsed_date_weeknum,
    count(1) AS "message_amount"

GROUP BY

    parsed_date_weeknum
ORDER BY
    parsed_date_weeknum,
    parsed_date_weekday
